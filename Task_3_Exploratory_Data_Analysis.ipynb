#Step 4: Exploratory Data Analysis

Visualizing the dataset as a whole using the pair plot

```python
# importing Seaborn library for interactive visuals
import seaborn as sns 
sns.pairplot(df)
```

```python
# finding the pairwise correlations between the columns and visualising using heatmaps
df.corr()
plt.figure(figsize=(10,5))
sns.heatmap(df.corr(), annot=True)
plt.show()
```

*Observation:*

*1.Sales and Profit are Moderately Correlated.*

*2.Quantity and Profit are less Moderately Correlated.*

*3.Discount and Profit are Negatively Correlated.*


__1.Ship Mode__

__Univariate Analysis__

```python
#plotting pairplot for Ship Mode
sns.pairplot(df, hue = 'Ship Mode')
```

```python
# find the count of unique values
df['Ship Mode'].value_counts()
```

```python
#Visualizing count plot for value counts of Ship Mode
sns.countplot(x=df['Ship Mode'])
```

__Bivariate Analysis: Profit vs Sales in different Ship Mode__

```python
df.groupby('Ship Mode')[['Profit','Sales']].sum().plot.bar(color=['black','blue'], figsize=(8,5))
plt.ylabel('Profit/Loss and sales')
plt.show()
```

*Observation 1: we can conclude that the profit and sales are the maximum in Standard Class and least in Same Day*

__2. Category__

__Univariate Analysis__

```python
#finding the value counts for Category
df['Category'].value_counts()
```

```python
#Visualiizing pie graph for Category
plt.figure(figsize = (5,5))
textprops = {"fontsize":15}
plt.title('Category')
plt.pie(df['Category'].value_counts(), labels = df['Category'].value_counts().index, autopct='%1.1f%%', textprops = textprops)
plt.show()
```

__Bivariate Analysis- Sales/Profit based on Category__

```python
# computing top categories in terms of sales from first 100 observations
top_category_s = df.groupby("Category").Sales.sum().nlargest(n=100)

# computing top categories in terms of profit from first 100 observations
top_category_p = df.groupby("Category").Profit.sum().nlargest(n=100)

# plotting to see it visually
plt.style.use('seaborn')
top_category_s.plot(kind = 'bar',figsize = (7,5),fontsize = 10)
top_category_p.plot(kind = 'bar',figsize = (7,5),fontsize = 10,color='red')

plt.xlabel('Category',fontsize = 15)
plt.ylabel('Total Sales/Profits',fontsize = 15)
plt.title("Top Category Sales vs Profit",fontsize = 15)
plt.show()
```

*Observation 2: From the displayed output we can visualize that the maximum profit and sales are in Technology category type.*

__3. Sub_Category__

__Univariate Analysis__

```python
df['Sub-Category'].value_counts()
```

```python
plt.figure(figsize=(10,8))
df['Sub-Category'].value_counts().plot.pie(autopct='%1.1f%%')
plt.show()
```

Maximum are from Binders, Paper, furnishings, Phones, storage, art, accessories and minimum from copiers, machines, suppliers.



__Bivariate Analysis- Sales/Profit based on Sub-Category:__

```python
# computing top sub-categories in terms of sales from first 100 observations
top_subcategory_s = df.groupby("Sub-Category").Sales.sum().nlargest(n = 100)

# computing top sub-categories in terms of profit from first 100 observations
top_subcategory_p = df.groupby("Sub-Category").Profit.sum().nlargest(n = 100)

# plotting to see it visually
plt.style.use('seaborn')
top_subcategory_s.plot(kind = 'bar',figsize = (10,5),fontsize = 14)
top_subcategory_p.plot(kind = 'bar',figsize = (10,5),fontsize = 14, color = 'red')

plt.xlabel('Sub-Category',fontsize = 15)
plt.ylabel('Total Sales/Profits',fontsize = 15)
plt.title("Top Sub-Category Sales vs Profit",fontsize = 15)
plt.show()
```

*Observation 3: From the demonstrated bar graph, we can clearly interpret that, Copiers have the highest number of Sales and Profits while tables have the lowest.*

```python
# A more detailed view
plt.figure(figsize=(10,8))
statewise = df.groupby(['Sub-Category'])['Profit'].sum().nlargest(50)
statewise.plot.barh() # h for horizontal
```

*The above graph clearly shows that Copiers and Phones have the highest sales and profit and tables has negative profit.*

__4. State__

__Univariate Analysis__

```python
df['State'].value_counts()
```

```python
plt.figure(figsize=(12,10))
sns.countplot(x='State',data=df,palette='rocket_r',order=df['State'].value_counts().index)
plt.xticks(rotation=90)
plt.show()
```

 *Highest number of buyers are from California and New York*

__Bivariate Analysis: Visualising the sales vs profit in different states__

```python
# computing top states in terms of sales from first 10 observations
top_states_s = df.groupby("State").Sales.sum().nlargest(n=10)

# computing top states in terms of profit from first 10 observations
top_states_p = df.groupby("State").Profit.sum().nlargest(n = 10)

plt.style.use('seaborn')
top_states_s.plot(kind = 'bar',figsize = (10,5),fontsize = 14)
top_states_p.plot(kind = 'bar',figsize = (10,5),fontsize = 14, color = 'Red')

plt.xlabel('States',fontsize = 15)
plt.ylabel('Total sales',fontsize = 15)
plt.title("Top 10 states Sales vs Profit",fontsize = 15)
plt.show()
```

*Observation 4:* 

*1.States like 'New York' and 'California' have the highest profit or sales.*

*2.Few states like North Dakota, Wyoming and West Virginia have less or no sales at all.*

*3.To generate more profit we have to increase our sales in Texas and New York where we have potential customers.*

__9. Checking the interdependency of Sales,Profits and Discounts__

```python
plt.style.use('seaborn')
df.plot(kind = "scatter",figsize = (10,5), x = "Sales", y= "Profit", c = "Discount", s = 20,fontsize = 16, colormap = "viridis")
plt.xlabel('Total Sales', fontsize = 15)
plt.ylabel('Total Profits',fontsize = 15)
plt.title("Interdependency of Sales,Profits and Discounts",fontsize = 16)

plt.show()
```

*Observation 9: So, the illustrated graph clearly implies that if we give more Discount on our products, sales increases but profit decreases.*
